#interpretación 

Aquí podemos observar la  base monetaria entre los periodos de 1997 y 2009; Esta la compone el dinero de las personas, el dinero bancario, dinero tangible o liquido.

Hay que recordar que la oferta monetaria incluye dinero fisico como dinero no tangible o liquido , por lo que no es lo mismo base monetaria a oferta monetaria 


En esta parte ingresaremos otra variable de nombre X en la cual incluiremos los datos que abarcan de 2007 a 2009 para poder comprender el comportamiento de la base monetaria y darle una interpretación a detalle

```python
start_date = '2007-01-01'
end_date = '2009-01-31'

# Convertir la columna 'fecha' a tipo de dato de fecha
df['fecha'] = pd.to_datetime(df['fecha'])

# Filtrar el DataFrame por el rango de fechas y seleccionar las columnas deseadas
x = df[(df['fecha'] >= start_date) & (df['fecha'] <= end_date)][['fecha', 'SF44043', 'SF44044']]
```

```python
plt.plot(x['fecha'],x['SF44043'])
```
